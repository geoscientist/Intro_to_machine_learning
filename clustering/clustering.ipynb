```python
import numpy as np
import math
from skimage.io import imread
from skimage import img_as_float
from skimage.viewer import ImageViewer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
%matplotlib inline
```

```python
image = imread('parrots.jpg')
img = mpimg.imread('parrots.jpg')
```

```python
image.shape
```

```python
plt.imshow(img)
```

```python
image_flt = img_as_float(image)
red = image_flt[:,:,0].flatten()
green = image_flt[:,:, 1].flatten()
blue = image_flt[:,:, 2].flatten()
```

```python
sns.distplot(red, color = "red")
sns.distplot(green, color = "green")
sns.distplot(blue, color = "blue")
```

```python
red = image_flt[:, :, 0].ravel()
green = image_flt[:, :, 1].ravel()
blue = image_flt[:, :, 2].ravel()

rgb = np.transpose(np.vstack((red, green, blue)))
```

```python
rgb
```

```python
clust = KMeans(init = 'k-means++', random_state=241)
```

```python
clust.fit(rgb)
```

```python
clust.labels_
```

```python
c = np.reshape(clust.labels_, (-1, 713))
c
```

```python
image_mn = np.copy(image_flt)
```

```python
for i in range(0, clust.n_clusters):
    image_mn[c == i] = clust.cluster_centers_[i]
```

```python
plt.imshow(image_mn)
```

```python
image_md = np.copy(image_flt)
```

```python
for i in range(0,clust.n_clusters):
    r_md = np.median(image_flt[:, :, 0][c == i])
    g_md = np.median(image_flt[:, :, 1][c == i])
    b_md = np.median(image_flt[:, :, 2][c == i])
    image_md[c == i] = [r_md, g_md, b_md]
plt.imshow(image_md)
```

```python
def PSNR(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    psnr = 10 * math.log10(np.max(img1) / mse)
    return psnr
```

```python
PSNR(image_flt, image_mn)
```

```python
for i in range(1, 21):
    clust = KMeans(n_clusters = i,
                init = 'k-means++', 
                random_state = 241)
    clust.fit(rgb)
    cl = clust.labels_
    colors_avg = clust.cluster_centers_
    c = np.reshape(cl, (-1, 713))
    img_new = np.copy(image_flt)
    for cluster in range(0, i):
        img_new[c == cluster] = colors_avg[cluster]
    print i, PSNR(image_flt, img_new)
```

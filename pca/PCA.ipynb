```python
import numpy as np
import pandas as pd
from sklearn import decomposition
```

```python
data = pd.read_csv('close_prices.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
pca = decomposition.PCA(n_components=10)
pca.fit(data.iloc[:, 1:31])
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
dj_data = pd.read_csv('djia_index.csv')
dj_data.head()
```

```python
data_tr = pca.transform(data.iloc[:, 1:31])
```

```python
cor = np.corrcoef(data_tr[:, 0], dj_data.ix[:, 1])
cor[0,1]
```

```python
companies = pd.DataFrame(data.columns[1:], pca.components_[0])
```

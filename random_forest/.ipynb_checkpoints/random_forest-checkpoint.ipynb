```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.metrics import r2_score
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('abalone.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
data.head()
```

```python
x = data.ix[:, 0:8]
y = data['Rings']
```

```python
kf = KFold(len(x), n_folds=5, shuffle=True, random_state=1)
```

```python
r_scores = pd.DataFrame(columns = ['N_trees', 'r_score'])
for i in range(1, 51):
    clf = RandomForestRegressor(n_estimators=i, random_state=1)
    r_list = cross_val_score(clf, x, y,'r2', cv = kf)
    r_scores.loc[i] = [i, np.mean(r_list)]
```

```python
r_scores
```

```python
sns.regplot(r_scores['N_trees'], r_scores['r_score'], fit_reg=False)
```
